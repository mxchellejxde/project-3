```python
# Dependencies and Setup
import pandas as pd

# File to Load 
data = "Resources/SPS Abundance - Population data and references for the Salmon Population Summary (SPS) Database Data.csv"

#read the excel
salmon_data = pd.read_csv(data)

salmon_data.head()
```

```python
salmon_data.columns
```

```python
#groupby year and species for salmon_data
salmon_year_species_df = salmon_data.groupby(['Species','Brood Year'], as_index=False)['Number Of Spawners'].sum()
salmon_year_species_df
```

```python
salmon_filtered = salmon_year_species_df[(salmon_year_species_df['Brood Year'] >= 2000) & (salmon_year_species_df['Species'] != "")].reset_index(drop=True)
salmon_filtered
```
